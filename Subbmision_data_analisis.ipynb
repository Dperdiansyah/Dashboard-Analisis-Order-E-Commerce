## Visualization & Explanatory Analysis

### Pertanyaan 1:

Bagaimana tren jumlah orderan setiap bulan selama periode data yang tersedia?

```python
monthly_orders_df = monthly_orders.sort_index().to_frame().reset_index()
monthly_orders_df.columns = ['order_month', 'total_orders']
monthly_orders_df['order_month'] = monthly_orders_df['order_month'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_orders_df, x='order_month', y='total_orders', marker='o')
plt.title('Jumlah orderan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah orderan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*Membuat grafik yang menampilkan jumlah orderan per bulan berdasarkan data yang tersedia*

### Pertanyaan 2:

Kota mana yang memberikan kontribusi orderan terbanyak?

```python
orders_customers_df = pd.merge(orders_df, customers_df, on='customer_id')

city_orders = orders_customers_df['customer_city'].value_counts().reset_index()
city_orders.columns = ['customer_city', 'total_orders']
top_cities = city_orders.head(5)

plt.figure(figsize=(12,6))
sns.barplot(data=top_cities, x='customer_city', y='total_orders')
plt.title('5 Kota dengan Jumlah Order Terbanyak')
plt.xlabel('Kota')
plt.ylabel('Jumlah Order')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Membuat grafik garis yang menampilkan 5 kota dengan orderan terbanyak

### Pertanyaan 3:

Metode pembayaran apa yang sering digunakan?

```python
payment_counts = order_payments_df['payment_type'].value_counts()

plt.figure(figsize=(9, 9))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Metode Pembayaran yang sering digunakan')
plt.axis('equal')
plt.show()
```

**Insight:**
- Terlihat bagaimana tren jumlah orderan berubah dari bulan ke bulan.
  Dari sini, bisa melihat apakah ada musim atau periode tertentu di mana jumlah orderan meningkat atau menurun. Misalnya, apakah ada lonjakan orderan pada bulan-bulan tertentu seperti musim liburan, promosi, atau event khusus
- Menunjukkan kota-kota dengan tingkat orderan tertinggi
  Dari grafik ini, bisa melihat konsentrasi orderan di kota-kota tertentu. Jika sebagian besar order berasal dari beberapa kota besar, ini bisa menunjukkan bahwa bisnis lebih berkembang di wilayah tersebut.
- Mayoritas pelanggan menggunakan kartu kredit sebagai metode pembayaran
